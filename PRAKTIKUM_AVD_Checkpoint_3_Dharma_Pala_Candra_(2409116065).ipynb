#**DATASET**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = "/content/katalog_gempa_cleaned.csv"
df = pd.read_csv(path)
df
```

#**VISUALISASI DATA**

Hubungan antara magnitudo dan kedalaman

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['depth'], df['mag'], c='blue', alpha=0.5)
plt.title('Hubungan antara magnitudo dan kedalaman')
plt.xlabel('Kedalaman')
plt.ylabel('Magnitudo')
plt.show()
```

Banyaknya gempa pada tahun 2008-2023

```python
df.value_counts("Keterangan")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Keterangan'].value_counts().index, df['Keterangan'].value_counts().values, color='Purple')
plt.title('Banyaknya gempa pada tahun 2008-2023')
plt.xlabel('Keterangan gempa')
plt.ylabel('Jumlah gempa')
plt.xticks(rotation=45)
plt.show()
```

Lokasi gempa melalui koordinat

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['lon'], df['lat'], c='red', alpha=0.5)
plt.title('Lokasi gempa melalui koordinat')
plt.xlabel('Lintang')
plt.ylabel('Bujur')
plt.show()
```

Banyaknya gempa pada daerah tertentu

```python
plt.figure(figsize=(16, 8))
plt.bar(df['remark'].value_counts().index, df['remark'].value_counts().values, color='Purple')
plt.title('Banyaknya gempa pada daerah tertentu')
plt.xlabel('Daerah gempa')
plt.ylabel('Jumlah gempa')
plt.xticks(rotation=90)
plt.show()
```

#**Analisi dan Tindakan**

Hasil visualisasi data di atas menunjukkan beberapa pola menarik terkait gempa bumi. Pertama, terdapat hubungan antara magnitudo dan kedalaman gempa, di mana semakin dalam sumber gempa, semakin bervariasi pula magnitudonya. Hal ini menunjukkan bahwa gempa dengan kedalaman yang lebih dalam cenderung memiliki kekuatan yang lebih beragam.

Pada grafik yang menunjukkan banyaknya gempa dari tahun 2008 hingga 2023, terlihat bahwa gempa dengan skala ringan (magnitudo kecil) merupakan jenis gempa yang paling sering terjadi. Selain itu, terdapat pola yang jelas: semakin besar magnitudo gempa, semakin jarang gempa tersebut terjadi. Ini mengindikasikan bahwa gempa berkekuatan besar merupakan kejadian yang lebih langka dibandingkan gempa ringan.

Sementara itu, dari grafik lokasi gempa yang ditampilkan melalui koordinat, terlihat bahwa sebagian besar gempa terkonsentrasi di sekitar wilayah Sumatera, Jawa, Sulawesi, dan Maluku. Wilayah-wilayah ini merupakan bagian dari Ring of Fire (Cincin Api Pasifik), yang dikenal sebagai area dengan aktivitas seismik dan vulkanik yang tinggi.

Pada grafik yang menunjukkan banyaknya gempa di daerah tertentu, wilayah Minahasa Peninsula - Sulawesi menonjol sebagai daerah yang paling sering mengalami gempa. Hal ini kemungkinan besar disebabkan oleh lokasinya yang berada di zona subduksi, di mana lempeng tektonik saling bertumbukan, sehingga memicu aktivitas seismik yang intens.

Tindakan Yang bisa diambil Berdasarkan Analisis diatas :
*  Peningkatan Sistem Peringatan Dini: Mengingat tingginya aktivitas gempa di wilayah-wilayah seperti Sumatera, Jawa, Sulawesi, dan Maluku, pemerintah dan lembaga terkait perlu memperkuat sistem peringatan dini gempa dan tsunami. Hal ini dapat membantu masyarakat untuk lebih siap menghadapi bencana.

* Edukasi dan Pelatihan Masyarakat: Masyarakat di daerah rawan gempa perlu diberikan edukasi tentang langkah-langkah mitigasi bencana, seperti cara menyelamatkan diri saat gempa terjadi dan persiapan menghadapi gempa besar.

* Pembangunan Infrastruktur Tahan Gempa: Di daerah seperti Minahasa Peninsula - Sulawesi, pembangunan infrastruktur harus memenuhi standar ketahanan gempa untuk mengurangi risiko kerusakan dan korban jiwa.

* Pemantauan Aktivitas Seismik Secara Intensif: Lembaga geologi dan vulkanologi perlu terus memantau aktivitas seismik di daerah rawan gempa, terutama di zona subduksi, untuk memberikan informasi yang akurat dan cepat kepada masyarakat.
